```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

Reading CSV file

```python
df = pd.read_csv('population_by_age_group.csv')
df
```

```python
df.shape
```

```python
df.size
```

```python
top_10 = df.sort_values(by='Total', ascending=False).head(10)
top_10
```

```python
plt.figure(figsize=(10,6))
plt.bar(top_10['Country'], top_10['Total'])  # Use DataFrame columns here
plt.xticks(rotation=45)
plt.xlabel('Country')
plt.ylabel('Total Population')
plt.title('Top 10 Countries by Total Population')
plt.tight_layout()
plt.show()
```

```python
top_countries = df.sort_values(by='Total', ascending=False).head(10)
top_countries
```

```python
age_groups = ['0-4 years', '5-14 years', '15-24 years', '25-64 years', '65+']
x = np.arange(len(top_countries['Country'])) 
width = 0.15  # width of each bar

plt.figure(figsize=(14, 6))

for i, age_group in enumerate(age_groups):
    plt.bar(x + (i - 2) * width, top_countries[age_group], width, label=age_group)

plt.xticks(x, top_countries['Country'], rotation=45)
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Population Distribution by Age Group')
plt.legend()
plt.tight_layout()
plt.show()
```
